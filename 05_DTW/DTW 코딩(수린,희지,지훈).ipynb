```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn import metrics
from sklearn.metrics import *
from tqdm import tqdm
from tslearn.clustering import TimeSeriesKMeans
from yellowbrick.cluster import SilhouetteVisualizer
from matplotlib import rc
import warnings
# warnings.filterwarnings('ignore')
# rc('font', family='NanumBarunGothic')
```

```python
distortions = []

for i in tqdm(range(2, 10)):
    model = TimeSeriesKMeans(n_clusters=i, metric="dtw", max_iter=10, random_state=42, n_jobs=-1)
    model.fit(data1111.drop('감정', axis=1))
    distortions.append(model.inertia_)

fig = plt.figure(figsize=(10, 5))
plt.plot(range(2, 10), distortions)
plt.grid(True)
plt.title('Elbow curve')
plt.show()
```

```python
from tslearn.clustering import silhouette_score

DB_vals = []
silhouette_vals = []

for i in tqdm(range(2,7)):
  model = TimeSeriesKMeans(n_clusters=i, metric="dtw", max_iter=10, random_state=42, n_jobs=-1)
  pred = model.fit_predict(data1111.drop('감정', axis=1))
  DB_vals.append(np.mean(metrics.davies_bouldin_score(data1111.drop('감정', axis=1), pred)))
  silhouette_vals.append(np.mean(silhouette_score(data1111.drop('감정', axis=1), pred, metric='dtw')))

plt.plot(range(2,7), silhouette_vals, marker='o', label='Silhouette Index')
plt.plot(range(2,7), DB_vals, marker='o', label='DB Index')

plt.legend()
plt.xlabel('Number of clusters')
plt.ylabel('Value')
plt.show()
```

```python
def plot_Silhouette(data,  k, text):
    model = TimeSeriesKMeans(n_clusters=k, metric="dtw", max_iter=10, random_state=42)
    model.fit(data)

    fig = plt.figure()
    fig.set_size_inches(15, 5)
    visualizer = SilhouetteVisualizer(model, colors='yellowbrick')
    visualizer.fit(data)

    visualizer.ax.set_title(text,fontsize=15, weight='bold')
    visualizer.ax.set_xlabel("Silhouette Value",fontsize=15, weight='bold')
    visualizer.ax.set_ylabel("군집",fontsize=15, weight='bold')
    visualizer.ax.set_yticks([])

    plt.show()
```

```python
def plot_Silhouette(data,  k, text):
    model = TimeSeriesKMeans(n_clusters=k, metric="dtw", max_iter=10, random_state=42)
    model.fit(data)

    fig = plt.figure()
    fig.set_size_inches(20, 10)
    visualizer = SilhouetteVisualizer(model, colors='yellowbrick')
    visualizer.fit(data)

    visualizer.ax.set_title(text,fontsize=15, weight='bold')
    visualizer.ax.set_xlabel("Silhouette Value",fontsize=15, weight='bold')
    visualizer.ax.set_ylabel("군집",fontsize=15, weight='bold')
    visualizer.ax.set_yticks([])

    plt.show()
```

```python
model = TimeSeriesKMeans(n_clusters=3, metric="dtw", max_iter=10, random_state=42, n_jobs=-1)
clus = model.fit_predict(data1111.drop('감정', axis=1))

clustering_result = pd.DataFrame({'감정' : data1111.감정, '군집' : clus})
```

```python
clustering_result.to_csv('감정별 클러스터링 결과.csv', index=False)
```
